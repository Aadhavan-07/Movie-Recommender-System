```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('movie-dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('movie-dataset/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(3)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(3)
```

```python
movies.isnull().sum()
```

```python
movies.shape
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.isnull().sum()
```

```python
import ast
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(3)
```

```python
movies['keywords'].iloc[0]
```

```python
movies['genres'].iloc[0]
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head(3)
```

```python
def convert_crew(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
movies.head(3)
```

```python
movies.rename(columns = {'crew':'director'}, inplace = True)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head(3)
```

```python
movies['overview'].iloc[3]
```

```python
#Sam Worthington
#SamWorthington

def remove_space(text):
    l=[]
    for i in text:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['director'] = movies['director'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['director']
```

```python
movies.head(3)
```

```python
import requests
def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=6ae70443ac46a1f9fa2f87ef818d81d2&language=en-US".format(movie_id)
    data = requests.get(url).json()
    poster_path = data['poster_path']
    full_path = "http://image.tmdb.org/t/p/w500" + poster_path
    return full_path
```

```python
new_movies = movies[['movie_id','title','tags']]
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(x))
new_movies.head()
```

```python
import nltk
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem(text):
    T=[]
    for i in text.split():
        T.append(ps.stem(i))
    return " ".join(T)  
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
new_movies.head()
```

```python
new_movies['tags'].iloc[5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python
vector[4]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse = True, key= lambda x:x[1])
    for i in distances[1:6]:
        print(new_movies['title'].iloc[i[0]])
```
