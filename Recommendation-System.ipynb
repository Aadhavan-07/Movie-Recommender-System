```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('movie-dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('movie-dataset/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(3)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(3)
```

```python
movies.isnull().sum()
```

```python
movies.shape
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['genres'].iloc[0]
```

```python
movies['keywords'].iloc[0]
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(3)
```

```python
def convert_crew(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head(3)
```

```python
movies.rename(columns = {'crew':'director'}, inplace = True)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head(3)
```

```python
movies['overview'].iloc[3]
```

```python
#Sam Worthington
#SamWorthington

def remove_space(text):
    l=[]
    for i in text:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['director'] = movies['director'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['director']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'].iloc[3]
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    for i in text.split():
        T.append(ps.stem(i))
    return " ".join(T)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[2]
```

```python
vector.shape
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse = True, key= lambda x:x[1])
    for i in distances[1:6]:
        print(new_df['title'].iloc[i[0]])
```

```python
recommend("Spider-Man")
```
